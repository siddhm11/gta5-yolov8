```python
import cv2
import os

# Path to the downloaded video
video_path = '/content/video2.mp4'

# Create a directory to save the frames
frames_dir = '/content/video2_frames'
os.makedirs(frames_dir, exist_ok=True)

# Open the video file
video_capture = cv2.VideoCapture(video_path)

# Initialize frame count
frame_count = 0
saved_frame_count = 0

# Loop through the video and save every 200th frame
while True:
    success, frame = video_capture.read()
    if not success:
        break

    # Save every 200th frame
    if frame_count % 200 == 0:
        frame_filename = os.path.join(frames_dir, f'frame_{saved_frame_count:05d}.jpg')
        cv2.imwrite(frame_filename, frame)
        saved_frame_count += 1

    frame_count += 1

# Release the video capture object
video_capture.release()

print(f"Extracted {saved_frame_count} frames (every 200th frame) to {frames_dir}")
```

```python

print("Training started...")

# Train the model with specified parameters
model.train(
    data="/content/gta5data/data.yaml",  # Path to your dataset configuration file
    epochs=15,                              # Number of epochs
    imgsz=416,                              # Image size
    batch=8,                                # Batch size
    augment=True,                           # Enable data augmentation (set to False if not needed)
    device=None,                            # Use None for CPU, or specify GPU (e.g., '0' for first GPU)
    workers=8,                              # Number of data loader workers
    project=None,                           # Project name
    name='train3',                          # Name of the training run
    save=True,                              # Save checkpoints
    verbose=True,                           # Print training details
    rect=False,                             # Use rectangular training (False for default)
    multi_scale=False,                      # Use multi-scale training (False for default)
    conf=0.01,                              # Confidence threshold for detection
    iou=0.7,                                # IoU threshold for NMS
    max_det=300,                            # Maximum number of detections per image
    optimizer='AdamW',                      # Optimizer to use
    lr0=0.001667,                          # Initial learning rate
    lrf=0.01,                               # Final learning rate (lr0 * lrf)
    momentum=0.9,                           # Momentum
    weight_decay=0.0005,                    # Weight decay
    warmup_epochs=3,                        # Number of warmup epochs
    warmup_momentum=0.8,                    # Warmup momentum
    warmup_bias_lr=0.1,                     # Warmup bias learning rate
    box=7.5,                                # Box loss gain
    cls=0.5,                                # Classification loss gain
    dfl=1.5,                                # Distribution Focal Loss gain
    pose=12.0,                              # Pose loss gain
    kobj=1.0,                               # Keypoint object loss gain
    label_smoothing=0.0,                    # Label smoothing
    nbs=64,                                 # Number of batch size for scaling
    hsv_h=0.015,                            # HSV hue augmentation
    hsv_s=0.7,                              # HSV saturation augmentation
    hsv_v=0.4,                              # HSV value augmentation
    degrees=0.0,                            # Degree of rotation augmentation
    translate=0.1,                          # Translation augmentation
    scale=0.5,                              # Scaling augmentation
    shear=0.0,                              # Shearing augmentation
    perspective=0.0,                        # Perspective transformation augmentation
    flipud=0.0,                             # Flip up-down augmentation
    fliplr=0.5,                             # Flip left-right augmentation
    bgr=0.0,                                # BGR color augmentation
    mosaic=1.0,                             # Mosaic augmentation
    mixup=0.0,                              # Mixup augmentation
    copy_paste=0.0,                         # Copy-paste augmentation
    auto_augment='randaugment',             # AutoAugment policy
    erasing=0.4,                            # Random erasing augmentation
    crop_fraction=1.0                       # Fraction of images to crop
)

print("Training ended...")
```
